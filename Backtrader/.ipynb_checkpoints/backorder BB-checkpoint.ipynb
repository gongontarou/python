```python
%%time
# â†‘ã‚»ãƒ«ã®å‡¦ç†æ™‚é–“ã‚’è¨ˆç®—ã€€ã‚»ãƒ«ã®æœ€åˆã«å˜ç‹¬ã§æ›¸ã

import os # ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒªä½œæˆã€ãƒ‘ã‚¹çµåˆã€ãƒ•ã‚¡ã‚¤ãƒ«å‰Šé™¤
import pandas as pd # pandasãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã‚’ä½¿ç”¨
import datetime

#dir_name = 'hst_20191128' # ãƒ’ã‚¹ãƒˆãƒªã‚«ãƒ«ãƒ‡ãƒ¼ã‚¿ã®ãƒ•ã‚©ãƒ«ãƒ€å

#input_csv = os.path.join(os.getcwd(), dir_name, 'USDJPY_20190930.csv') # csvãƒ•ã‚¡ã‚¤ãƒ«ã®ãƒ•ãƒ«ãƒ‘ã‚¹

df = pd.read_csv('tsd.csv') # csvãƒ•ã‚¡ã‚¤ãƒ«ã‚’Pandasãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã«èª­ã¿è¾¼ã‚€

#print(df.tail())

#æ—¥æ™‚åˆ—ã‚’datatimeåž‹ã«ã—ã¦ã‚¤ãƒ³ãƒ‡ãƒƒã‚¯ã‚¹ã«ã—ã¦ã€å…ƒã®åˆ—ã¯å‰Šé™¤ã™ã‚‹
df = df.set_index(pd.to_datetime(df['Date'])).drop('Date', axis=1) 

import backtrader as bt # Backtrader
import backtrader.feeds as btfeed # ãƒ‡ãƒ¼ã‚¿å¤‰æ›


data = btfeed.PandasData(dataname=df) # Pandasã®ãƒ‡ãƒ¼ã‚¿ã‚’Backtraderã®å½¢å¼ã«å¤‰æ›ã™ã‚‹

class myStrategy(bt.Strategy): # ã‚¹ãƒˆãƒ©ãƒ†ã‚¸ãƒ¼

    def log(self, txt, dt=None, doprint=False): # ãƒ­ã‚°å‡ºåŠ›ç”¨ã®ãƒ¡ã‚½ãƒƒãƒ‰
        if doprint:
            print('{0:%Y-%m-%d %H:%M:%S}, {1}'.format(
                dt or self.datas[0].datetime.datetime(0),
                txt
            ))

    def __init__(self):
        self.boll = bt.indicators.BollingerBands(self.data, period=25)
 
    def next(self):

        if not self.position:
            if self.data.close > self.boll.lines.top:
                self.sell()
        else:        
            if self.data.close < self.boll.lines.bot:
                self.buy()
 
            
    def notify_order(self, order): # æ³¨æ–‡ã®ã‚¹ãƒ†ãƒ¼ã‚¿ã‚¹ã®å¤‰æ›´ã‚’é€šçŸ¥ã™ã‚‹
        if order.status in [order.Submitted, order.Accepted]: # æ³¨æ–‡ã®çŠ¶æ…‹ãŒé€ä¿¡æ¸ˆorå—ç†æ¸ˆã®å ´åˆ
            return # ä½•ã‚‚ã—ãªã„
        
        if order.status in [order.Completed]: # æ³¨æ–‡ã®çŠ¶æ…‹ãŒå®Œäº†æ¸ˆã®å ´åˆ
            if order.isbuy(): # è²·ã„æ³¨æ–‡ã®å ´åˆ
                self.log('è²·ã„ç´„å®š, å–å¼•é‡:{0:.2f}, ä¾¡æ ¼:{1:.2f}, å–å¼•é¡:{2:.2f}, æ‰‹æ•°æ–™:{3:.2f}'.format(
                        order.executed.size, # å–å¼•é‡
                        order.executed.price, # ä¾¡æ ¼
                        order.executed.value, # å–å¼•é¡
                        order.executed.comm # æ‰‹æ•°æ–™
                    ), 
                    dt=bt.num2date(order.executed.dt), # ç´„å®šã®æ—¥æ™‚ã‚’datetimeåž‹ã«å¤‰æ›
                    #doprint=True # Trueã®å ´åˆå‡ºåŠ›
                )
            elif order.issell(): # å£²ã‚Šæ³¨æ–‡ã®å ´åˆ
                self.log('å£²ã‚Šç´„å®š, å–å¼•é‡:{0:.2f}, ä¾¡æ ¼:{1:.2f}, å–å¼•é¡:{2:.2f}, æ‰‹æ•°æ–™:{3:.2f}'.format(
                        order.executed.size, # å–å¼•é‡
                        order.executed.price, # ä¾¡æ ¼
                        order.executed.value, # å–å¼•é¡
                        order.executed.comm # æ‰‹æ•°æ–™
                    ), 
                    dt=bt.num2date(order.executed.dt), # ç´„å®šã®æ—¥æ™‚ã‚’datetimeåž‹ã«å¤‰æ›
                    #doprint=True # Trueã®å ´åˆãƒ­ã‚°ã‚’å‡ºåŠ›ã™ã‚‹
                )
                
        # æ³¨æ–‡ã®çŠ¶æ…‹ãŒã‚­ãƒ£ãƒ³ã‚»ãƒ«æ¸ˆãƒ»ãƒžãƒ¼ã‚¸ãƒ³ã‚³ãƒ¼ãƒ«ï¼ˆè¨¼æ‹ é‡‘ä¸è¶³ï¼‰ãƒ»æ‹’å¦æ¸ˆã®å ´åˆ
        elif order.status in [order.Canceled, order.Margin, order.Rejected]:
            self.log('æ³¨æ–‡ ã‚­ãƒ£ãƒ³ã‚»ãƒ«ãƒ»ãƒžãƒ¼ã‚¸ãƒ³ã‚³ãƒ¼ãƒ«ï¼ˆè¨¼æ‹ é‡‘ä¸è¶³ï¼‰ãƒ»æ‹’å¦',
                #doprint=True
            )

    def notify_trade(self, trade): # å–å¼•ã®é–‹å§‹/æ›´æ–°/çµ‚äº†ã‚’é€šçŸ¥ã™ã‚‹
        if trade.isclosed: # ãƒˆãƒ¬ãƒ¼ãƒ‰ãŒå®Œäº†ã—ãŸå ´åˆ
            self.log('å–å¼•æç›Š, ç·é¡:{0:.2f}, ç´”é¡:{1:.2f}'.format(
                    trade.pnl, # æç›Š
                    trade.pnlcomm # æ‰‹æ•°æ–™ã‚’å·®ã—å¼•ã„ãŸæç›Š
                ),
                #doprint=True # Trueã®å ´åˆãƒ­ã‚°ã‚’å‡ºåŠ›ã™ã‚‹
            )

# ãƒãƒƒã‚¯ãƒ†ã‚¹ãƒˆã®è¨­å®š
cerebro = bt.Cerebro() # Cerebroã‚¨ãƒ³ã‚¸ãƒ³ã‚’ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹åŒ–
cerebro.addstrategy(myStrategy) # ã‚¹ãƒˆãƒ©ãƒ†ã‚¸ãƒ¼ã‚’è¿½åŠ 
data = btfeed.PandasData(dataname=df) # Cerebroå½¢å¼ã«ãƒ‡ãƒ¼ã‚¿ã‚’å¤‰æ›
cerebro.adddata(data) # ãƒ‡ãƒ¼ã‚¿ã‚’Cerebroã‚¨ãƒ³ã‚¸ãƒ³ã«è¿½åŠ 
cerebro.broker.setcash(1000000.0) # æ‰€æŒé‡‘ã‚’è¨­å®š
cerebro.broker.setcommission(commission=0.0005) # æ‰‹æ•°æ–™ï¼ˆã‚¹ãƒ—ãƒ¬ãƒƒãƒ‰ï¼‰ã‚’0.05%ã«è¨­å®š
cerebro.addsizer(bt.sizers.PercentSizer, percents=50) # ãƒ‡ãƒ•ã‚©ãƒ«ãƒˆï¼ˆbuy/sellã§å–å¼•é‡ã‚’è¨­å®šã—ã¦ã„ãªã„æ™‚ï¼‰ã®å–å¼•é‡ã‚’æ‰€æŒé‡‘ã«å¯¾ã™ã‚‹å‰²åˆã§æŒ‡å®šã™ã‚‹
startcash = cerebro.broker.getvalue() # é–‹å§‹æ™‚ã®æ‰€æŒé‡‘
cerebro.broker.set_coc(True) # ç™ºæ³¨æ™‚ã®çµ‚å€¤ã§ç´„å®šã™ã‚‹

# è§£æžã®è¨­å®š
import backtrader.analyzers as btanalyzers # ãƒãƒƒã‚¯ãƒ†ã‚¹ãƒˆã®è§£æžç”¨ãƒ©ã‚¤ãƒ–ãƒ©ãƒª
cerebro.addanalyzer(btanalyzers.DrawDown, _name='myDrawDown') # ãƒ‰ãƒ­ãƒ¼ãƒ€ã‚¦ãƒ³
cerebro.addanalyzer(btanalyzers.SQN, _name='mySQN') # SQN
cerebro.addanalyzer(btanalyzers.TradeAnalyzer, _name='myTradeAnalyzer') # ãƒˆãƒ¬ãƒ¼ãƒ‰ã®å‹æ•—ç­‰ã®çµæžœ

thestrats = cerebro.run() # ãƒãƒƒã‚¯ãƒ†ã‚¹ãƒˆã‚’å®Ÿè¡Œ
thestrat = thestrats[0] # è§£æžçµæžœã®å–å¾—

# è©•ä¾¡å€¤ã®è¡¨ç¤º
print('Start                 :{0:%Y/%m/%d %H:%M:%S}'.format( # ãƒ’ã‚¹ãƒˆãƒªã‚«ãƒ«ãƒ‡ãƒ¼ã‚¿ã®é–‹å§‹æ—¥æ™‚
    pd.to_datetime(df.index.values[0])
))
print('End                   :{0:%Y/%m/%d %H:%M:%S}'.format (# ãƒ’ã‚¹ãƒˆãƒªã‚«ãƒ«ãƒ‡ãƒ¼ã‚¿ã®é–‹å§‹æ—¥æ™‚
    pd.to_datetime(df.index.values[-1])
))
print('Duration              :{0}'.format( # ãƒ’ã‚¹ãƒˆãƒªã‚«ãƒ«ãƒ‡ãƒ¼ã‚¿ã®æœŸé–“ã®é•·ã•
    pd.to_datetime(df.index.values[-1]) - pd.to_datetime(df.index.values[0])
))
print('Exposure[%]           :{0:.2f}'.format( # ãƒã‚¸ã‚·ãƒ§ãƒ³ã‚’æŒã£ã¦ã„ãŸæœŸé–“ã®å‰²åˆï¼ˆãƒã‚¸ã‚·ãƒ§ãƒ³ã‚’æŒã£ã¦ã„ãŸæœŸé–“Ã·å…¨æœŸé–“Ã—100ï¼‰
    thestrat.analyzers.myTradeAnalyzer.get_analysis().len.total / len(df) * 100
))
print('Equity Final[$]       :{0:.2f}'.format( # æ‰€æŒé‡‘ã®æœ€çµ‚å€¤ï¼ˆclosedã—ãŸå–å¼•ï¼‰
    startcash + thestrat.analyzers.myTradeAnalyzer.get_analysis().pnl.net.total
))  
print('Return[%]             :{0:.2f}'.format( # åˆ©ç›ŠçŽ‡=æç›ŠÃ·é–‹å§‹æ™‚æ‰€æŒé‡‘Ã—100
    thestrat.analyzers.myTradeAnalyzer.get_analysis().pnl.net.total / startcash * 100
))
print('Buy & Hold Return[%]  :{0:.2f}'.format( # ï¼ˆï¼ˆçµ‚äº†æ™‚ã®çµ‚å€¤ - é–‹å§‹æ™‚ã®çµ‚å€¤ï¼‰Ã· é–‹å§‹æ™‚ã®çµ‚å€¤ï¼‰ã®çµ¶å¯¾å€¤Ã—100
    (df['Close'][-1] - df['Close'][0]) / df['Close'][0] * 100
))
print('Max. Drawdown[%]      :{0:.2f}'.format( # æœ€å¤§ä¸‹è½çŽ‡
    thestrat.analyzers.myDrawDown.get_analysis().max.drawdown
))
print('Max. Drawdown Duration:{0}'.format( # æœ€å¤§ä¸‹è½æœŸé–“
    pd.Timedelta(minutes=thestrat.analyzers.myDrawDown.get_analysis().max.len)
))
print('Trades                :{0}'.format( # å–å¼•å›žæ•°
    thestrat.analyzers.myTradeAnalyzer.get_analysis().total.closed
))
winrate = ( # å‹çŽ‡
    thestrat.analyzers.myTradeAnalyzer.get_analysis().won.total 
    / thestrat.analyzers.myTradeAnalyzer.get_analysis().total.closed
)
lostrate = ( # æ•—çŽ‡
    thestrat.analyzers.myTradeAnalyzer.get_analysis().lost.total 
    / thestrat.analyzers.myTradeAnalyzer.get_analysis().total.closed
)
print('Win Rate[%]           :{0:.2f}'.format( # å‹çŽ‡=å‹ã¡å–å¼•å›žæ•°Ã·å…¨å–å¼•å›žæ•°Ã—100
    winrate * 100
))
print('Best Trade[%]         :{0:.2f}'.format( # 1å›žã®å–å¼•ã§ã®åˆ©ç›Šã®æœ€å¤§å€¤Ã·æ‰€æŒé‡‘Ã—100
    thestrat.analyzers.myTradeAnalyzer.get_analysis().won.pnl.max / startcash * 100
))
print('Worst Trade[%]        :{0:.2f}'.format( # 1å›žã®å–å¼•ã§ã®æå¤±ã®æœ€å¤§å€¤Ã·æ‰€æŒé‡‘Ã—100
    thestrat.analyzers.myTradeAnalyzer.get_analysis().lost.pnl.max / startcash * 100
))
print('Avg. Trade[%]         :{0:.2f}'.format( # æç›Šã®å¹³å‡å€¤Ã·æ‰€æŒé‡‘Ã—100
    thestrat.analyzers.myTradeAnalyzer.get_analysis().pnl.net.average / startcash * 100
))
print('Max. Trade Duration   :{0}'.format( # 1å›žã®å–å¼•ã§ã®æœ€é•·æœŸé–“
    pd.Timedelta(minutes=thestrat.analyzers.myTradeAnalyzer.get_analysis().len.max)
))
print('Avg. Trade Duration   :{0}'.format( # 1å›žã®å–å¼•ã§ã®å¹³å‡æœŸé–“
    pd.Timedelta(minutes=thestrat.analyzers.myTradeAnalyzer.get_analysis().len.average)
))
print('Expectancy[%]         :{0:.2f}'.format( # æœŸå¾…å€¤=å¹³å‡åˆ©ç›ŠÃ—å‹çŽ‡ï¼‹å¹³å‡æå¤±Ã—æ•—çŽ‡
    thestrat.analyzers.myTradeAnalyzer.get_analysis().won.pnl.average * winrate 
    + thestrat.analyzers.myTradeAnalyzer.get_analysis().lost.pnl.average * lostrate
))
print('SQN                   :{0:.2f}'.format( # SQN ã‚·ã‚¹ãƒ†ãƒ ã®è©•ä¾¡å€¤
    thestrat.analyzers.mySQN.get_analysis().sqn
))

# ã‚°ãƒ©ãƒ•ã®è¨­å®š
%matplotlib inline
# â†‘ã‚°ãƒ©ãƒ•ã‚’Notebookå†…ã«æç”»ã™ã‚‹

import matplotlib.pylab as pylab # ã‚°ãƒ©ãƒ•æç”»ç”¨ãƒ©ã‚¤ãƒ–ãƒ©ãƒª

pylab.rcParams['figure.figsize'] = 12, 8  # ã‚°ãƒ©ãƒ•ã®ã‚µã‚¤ã‚º
cerebro.plot(
    style = 'candle', # ãƒ­ã‚¦ã‚½ã‚¯è¡¨ç¤ºã«ã™ã‚‹
    barup = 'green', barupfill = False, # é™½ç·šã®è‰²ã€å¡—ã‚Šã¤ã¶ã—è¨­å®š
    bardown = 'red', bardownfill = False, # é™°ç·šã®è‰²ã€å¡—ã‚Šã¤ã¶ã—è¨­å®š
    fmt_x_data = '%Y-%m-%d %H:%M:%S' # æ™‚é–“è»¸ã®ãƒžã‚¦ã‚¹ã‚ªãƒ¼ãƒãƒ¼æ™‚ã®è¡¨ç¤ºãƒ•ã‚©ãƒ¼ãƒžãƒƒãƒˆ
)


```
